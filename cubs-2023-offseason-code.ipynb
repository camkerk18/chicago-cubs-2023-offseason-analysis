Starting Code:

df_bat = overall team batting data averages by year from 1871-2022

df_pitch = overall team pitching data averages by year from 1871-2022

df_batpitch = df_bat and df_pitch in one dataframe

df_bat_contemp = batting data of the contemporary era (1980 to Present)


```python
import pandas as pd
```

```python
df_bat = pd.read_csv("batting.csv")
df_bat.fillna(0)
df_bat
```

```python
df_pitch = pd.read_csv("pitching.csv")
df_pitch.fillna(0)
df_pitch
```

```python
df_batpitch = pd.merge(df_bat, df_pitch, how = "left", on = "Year")
df_batpitch
```

```python
df_bat_contemp = df_bat[df_bat.Year >= 1980]
df_bat_contemp
```
